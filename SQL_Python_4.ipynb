```python
import pandas as pd                     # Pandas is used for handling data tables (DataFrames)
import sqlite3                          # sqlite3 allows Python to connect to and manage SQLite databases
import prettytable                      # PrettyTable is used to format SQL query results nicely in Jupyter

prettytable.DEFAULT = 'DEFAULT'         #Changes internal attribute of prettytable.

%load_ext sql                           # Load the SQL extension for Jupyter/IPython.

# Conect ipython-sql Ã  BD (Essential to use %sql/%%sql)
%sql sqlite:///RealWorldData.db

# Read file
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv"
df = pd.read_csv(url)

# Conect to DB
con = sqlite3.connect("RealWorldData.db")

# Record CSV values in DB
df.to_sql("chicago_socioeconomic_data", con, if_exists='replace', index=False)

# Check if recorded
print("Tabela criada com sucesso com", len(df), "linhas.")
```

```python
#Get all table names from DB
%sql SELECT name FROM sqlite_master WHERE type='table';
```

```python
#Get all the columns
%sql SELECT name FROM pragma_table_info('chicago_socioeconomic_data');
```

```python
%sql SELECT * FROM chicago_socioeconomic_data limit 5;
```

```python
#Number of Elementary Schools 
%sql SELECT COUNT (*) FROM chicago_socioeconomic_data WHERE [Elementary, Middle, or High School]='ES';
```

```python
#Highest Safety Score
%sql SELECT MAX (SAFETY_SCORE) FROM chicago_socioeconomic_data;
```

```python
#Schools with the Highest Safety Score
%sql SELECT NAME_OF_SCHOOL FROM chicago_socioeconomic_data WHERE [SAFETY_SCORE]=99;
```

```python
#top 10 schools with the highest "Average Student Attendance"
%sql select Name_of_School, Average_Student_Attendance from chicago_socioeconomic_data order by Average_Student_Attendance desc nulls last limit 10;
```

```python
#list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance
%sql select Name_of_School, Average_Student_Attendance from chicago_socioeconomic_data order by Average_Student_Attendance nulls last limit 5;
```

```python
#Remove the '%' from th Average Student Attendance column - replace - 6
%sql select Name_of_School, REPLACE(Average_Student_Attendance, '%', '') from chicago_socioeconomic_data order by Average_Student_Attendance nulls last limit 5;
```
